```python
import os 
import pandas as pd
import numpy as np
from dotenv import load_dotenv
from sklearn.model_selection import train_test_split
```

```python
load_dotenv()
mod_train_path = os.getenv('mod_train_path')
mod_val_path = os.getenv('mod_val_path')
train_x_np_path  = os.getenv('train_x_np_path')
train_y_np_path = os.getenv('train_y_np_path')
val_x_np_path = os.getenv('val_x_np_path')
val_y_np_path = os.getenv('val_y_np_path')
```

```python
traindf = pd.read_csv(mod_train_path)
traindf.sample(frac=1)
traindf[:5]
```

```python
z = pd.get_dummies(traindf['Y'],dtype=float)
traindf.drop(['Y','Unnamed: 0'],inplace=True,axis=1)
X = np.stack(traindf['X'].values)
Y = z.values
```

```python
X[:5]
```

```python
Y[:5]
```

```python
np.save(train_x_np_path,X)
np.save(train_y_np_path,Y)
```

```python
valdf = pd.read_csv(mod_val_path)
valdf.sample(frac=1)
valdf[:5]
```

```python
z = pd.get_dummies(valdf['Y'],dtype=float)
valdf.drop(['Y','Unnamed: 0'],inplace=True,axis=1)
X = np.stack(valdf['X'].values)
Y = z.values
```

```python
np.save(val_x_np_path,X)
np.save(val_y_np_path,Y)
```
