```python
import os 
import pandas as pd
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from dotenv import load_dotenv
```

```python
load_dotenv()
raw_train_path = os.getenv('raw_train_path')
raw_val_path = os.getenv('raw_val_path')
mod_train_path = os.getenv('mod_train_path')
mod_val_path = os.getenv('mod_val_path')
```

```python
def Imgtransfrom(x):
    z = []
    for i in x:
        image = Image.open(i)
        reimage = image.resize((244,244))
        image_array = np.array(reimage)
        image_array = image_array/255.0
        z.append(image_array)
    return z
```

```python
train_x = []
trian_y = []

for fname in os.listdir(path=raw_train_path):
    file_path = os.path.join(raw_train_path, fname)
    i = 0
    for j in os.listdir(path=file_path):
        main_path = os.path.join(file_path,j)
        i+=1
        train_x.append(main_path)
        trian_y.append(fname)
        if i>100:
            break
```

```python
print(len(train_x))
print(len(trian_y))
```

```python
image_x = Imgtransfrom(x=train_x)
```

```python
traindf=pd.DataFrame({"X":image_x,"Y":trian_y})
traindf.to_csv(mod_train_path)
```

```python
val_x = []
val_y = [] 

for fname in os.listdir(path=raw_val_path):
    file_path = os.path.join(raw_val_path, fname)
    i = 0
    for j in os.listdir(path=file_path):
        main_path = os.path.join(file_path,j)
        i+=1
        val_x.append(main_path)
        val_y.append(fname)
        if i>50:
            break
```

```python
print(len(val_x))
print(len(val_y))
```

```python
val_image_x = Imgtransfrom(x=val_x)
```

```python
valdf=pd.DataFrame({"X":val_image_x,"Y":val_y})
valdf.to_csv(mod_val_path)
```
