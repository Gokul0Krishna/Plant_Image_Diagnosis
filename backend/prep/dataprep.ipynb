```python
import os 
import pandas as pd
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from dotenv import load_dotenv
```

```python
load_dotenv()
raw_train_path = os.getenv('raw_train_path')
raw_val_path = os.getenv('raw_val_path')
train_x_np_path  = os.getenv('train_x_np_path')
train_y_np_path = os.getenv('train_y_np_path')
val_x_np_path = os.getenv('val_x_np_path')
val_y_np_path = os.getenv('val_y_np_path')
```

```python
def Imgtransfrom(x):
    z = []
    for i in x:
        image = Image.open(i).convert('RGB')
        reimage = image.resize((244,244))
        image_array = np.array(reimage,dtype=np.float32)
        image_array = image_array/255.0
        z.append(image_array)
    return z
```

```python
train_x = []
trian_y = []

for fname in os.listdir(path=raw_train_path):
    file_path = os.path.join(raw_train_path, fname)
    i = 0
    for j in os.listdir(path=file_path):
        main_path = os.path.join(file_path,j)
        i+=1
        train_x.append(main_path)
        trian_y.append(fname)
        if i>100:
            break
```

```python
print(len(train_x))
print(len(trian_y))
```

```python
image_x = Imgtransfrom(x=train_x)
```

```python
series = pd.Series({"Y":trian_y})
z = pd.get_dummies(series['Y'],dtype=float)
X = np.stack(image_x)
Y = z.values
```

```python
Y.shape
```

```python
np.save(train_x_np_path,X)
np.save(train_y_np_path,Y)
```

```python
val_x = []
val_y = [] 

for fname in os.listdir(path=raw_val_path):
    file_path = os.path.join(raw_val_path, fname)
    i = 0
    for j in os.listdir(path=file_path):
        main_path = os.path.join(file_path,j)
        i+=1
        val_x.append(main_path)
        val_y.append(fname)
        if i>50:
            break
```

```python
print(len(val_x))
print(len(val_y))
```

```python
val_image_x = Imgtransfrom(x=val_x)
```

```python
series = pd.Series({"Y":val_y})
z = pd.get_dummies(series['Y'],dtype=float)
X = np.stack(val_image_x)
Y = z.values
```

```python
Y.shape
```

```python
np.save(val_x_np_path,X)
np.save(val_y_np_path,Y)
```
