```python
import fitz
import re
import os
import pandas as pd
from dotenv import load_dotenv
```

```python
load_dotenv()

main_pdf_path = os.getenv('main_pdf_path')
series_chunk_path = os.getenv('series_chunk_path')
```

```python
doc = fitz.open(main_pdf_path)
text = ''
for page in doc:
    text += page.get_text()
```

```python
text = re.sub(r'\n+', '\n', text)
text = re.sub(r'\s+', ' ', text)

cleaned_text = text.strip()
```

```python
chunks = []
words = cleaned_text.split()
for i in range(0,len(words),100):
    temp = ''.join(words[i:i+100])
    chunks.append(temp)
```

```python
df = pd.Series(chunks)
df.to_csv(series_chunk_path,index=False)
```
