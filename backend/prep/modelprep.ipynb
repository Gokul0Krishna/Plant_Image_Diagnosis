```python
import pandas as pd
import numpy as np
from tensorflow.keras import layers, models
from sklearn.metrics import accuracy_score,precision_score,confusion_matrix
import os
from dotenv import load_dotenv
```

```python
load_dotenv()
train_x_np_path  = os.getenv('train_x_np_path')
train_y_np_path = os.getenv('train_y_np_path')
val_x_np_path = os.getenv('val_x_np_path')
val_y_np_path = os.getenv('val_y_np_path')
```

```python
trainX = np.load(train_x_np_path)
trainY = np.load(train_y_np_path)
valX = np.load(val_x_np_path)
valY = np.load(val_y_np_path)
```

```python
model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=trainX.shape[1:]),
    layers.MaxPooling2D(2,2),

    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),

    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(trainY.shape[1], activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(trainX, trainY, epochs=10)
```

```python
model.save_weights("leaf_model.weights.h5")
```
